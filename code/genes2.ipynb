##üçä **Building a simple Python blockchain that embedds Secure Hash Algorithms**

```python
class GeekCoinBlock:

    def __init__(self, previous, transaction_list):

        self.previous = previous
        self.transaction_list = transaction_list

        self.block_data = f"{' - '.join(transaction_list)} - {previous}"
        self.block_hash = hashlib.sha256(self.block_data.encode()).hexdigest()
```

```python
class Blockchain:
    def __init__(self):
        self.chain = []                       #chain is where all blocks are recorded
        self.generate_first_block()

    def generate_first_block(self):               #also called genesis
        self.chain.append(GeekCoinBlock("0", ['First Block']))

    def create_block_from_transaction(self, transaction_list):
        previous_block_hash = self.last_block.block_hash
        self.chain.append(GeekCoinBlock(previous_block_hash, transaction_list))

    def display(self):
        for i in range(len(self.chain)):
            print(f"Data {i + 1}: {self.chain[i].block_data}")
            print(f"Hash {i + 1}: {self.chain[i].block_hash}\n")

    @property
    def last_block(self):
        return self.chain[-1]
```

##üçä **Comparing SHA algorithms with each other**

## SHA-256

```python
%%time

for i in x_train:
  hashedval256 = hashlib.sha256(i.encode())
  hex256 = hashedval256.hexdigest()
  print(hex256)
```

#### ‚è≤ Time taken is a total of 2.41 seconds.



### SHA-512

```python
%%time

for i in x_train:
  hashedval512 = hashlib.sha512(i.encode())
  hex512 = hashedval512.hexdigest()
  print(hex512)
```

#### ‚è≤ Time taken is a total of 2.04 seconds.

##SHA-224

```python
%%time
for i in x_train:
  hashedval224 = hashlib.sha224(i.encode())
  hex224 = hashedval224.hexdigest()
  print(hex224)
```

#### ‚è≤ Time taken is a total of 2.14 seconds.

```python
##comparing them and graphing them

names = ['SHA-256','SHA-512','SHA-224']
timings = [2.41,2.04,2.22]
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(names, timings)),
               columns =['Encoding Algorithms', 'Time taken'])
df
```
